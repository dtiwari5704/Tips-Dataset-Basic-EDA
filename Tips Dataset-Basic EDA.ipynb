# Tips Dataset


`Data Features:`

**total_bill**: a numeric vector, the bill amount (dollars)


**tip**: a numeric vector, the tip amount (dollars)


**sex**: a factor with levels Female Male, gender of the payer of the bill


**smoker**: a factor with levels No Yes, whether the party included smokers


**day**: a factor with levels Friday Saturday Sunday Thursday, day of the week


**time**: a factor with levels Day Night, rough time of day


**size**: a numeric vector, number of people in a table/Party

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#3D59AB;">1.1 Import the libraries

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
```

<a id = "link2"></a> 
<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#3D59AB;">1.2 Get the Dataset. Check info and shape. Display first 5 rows.

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

```python
tips.describe().T
```

<a id = "link3"></a> 
<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#3D59AB;">1.3 Understand the distribution of 'tip' and 'total_bill' using different plots

```python
sns.distplot(tips['tip'],color='red')
plt.show()
sns.boxplot(tips['tip'],orient='v')
plt.show()
```

```python
sns.distplot(tips['total_bill'],color='lime')
plt.show()
sns.boxplot(tips['total_bill'],orient='v')
plt.show()
```

##### Observation
* Both 'tip' and 'total_bill' are skewed towards right
* 50 % people paying less than 20 dollars for overall bill and 50 % people paying less than 3 dollars as tip

<a id = "link4"></a> 
<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#3D59AB;">1.4 Relationship between 'total_bill' and 'tip'

```python
sns.jointplot(tips['total_bill'], tips['tip'],color='blue')
plt.show()
```

```python
sns.jointplot(tips['total_bill'], tips['tip'], kind = "hex",color='red')
plt.show()
```

```python
sns.pairplot(tips[['total_bill','tip']])
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tip)
plt.show()
```

<a id = "link5"></a> 
<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#3D59AB;">1.5 Make plots to check distribution of tip with different gender

```python
sns.stripplot(tips['sex'], tips['tip'])
plt.show()
```

```python
sns.boxplot(tips['sex'], tips['tip'])
plt.show()
```

##### Observation:
* Median value of tips is approx same for both male and female
* male have more outliers giving tips than female

<a id = "link6"></a> 
<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#3D59AB;">1.6 Draw relationship between 'total_bill', 'tip' and 'sex'

```python
sns.lmplot(x='total_bill',y='tip',data=tips,fit_reg=False,hue='sex')
plt.show()
```

* higher billing amount and tip are from males

<a id = "link7"></a> 
<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#3D59AB;">1.7 Which day bring in higher bill amount? 

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False)
plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.boxplot(y='total_bill',x='day',data=tips)
plt.show()
```

* Number of customers and total bills are high on Sat and Sun.

<a id = "link8"></a> 
<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#3D59AB;">1.8 Find out count of total males and females visiting each day

```python
sns.countplot(tips['day'], hue = tips['sex'])
plt.show()
```

<a id = "link9"></a> 
<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#3D59AB;">1.9 At what time of which day does the restaurant get higher amount of bill

```python
sns.lmplot(x='total_bill',y='tip',data=tips,fit_reg=False,hue='day',row='sex',col='time')
plt.show()
```

* Male and female spend more at Dinners on Sat and Sun

<a id = "link11"></a> 
<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#3D59AB;">1.11 Draw relationship between 'total_bill', 'tip' and 'smokers'

```python
sns.countplot(tips['smoker'])
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,fit_reg=False,hue='sex',col='smoker')
plt.show()
```

* Non smoker show a more prominent linear relationship between total_bill and tip compared to smokers

* Highest tip was given by a smokers (Outlier)
